```
#File Declaration 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets as dat
```

```
iris = dat.load_iris()
iris_df = pd.DataFrame(iris.data, columns=['sepal_len','sepal_width','petal_len','petal_width'])
print(iris_df.head())
```

```
#Joint Probability of Petal Length and Sepal Length
#values between 0 and 1
joint_prob = pd.crosstab(index=iris_df['petal_len'], columns=iris_df['sepal_len'], normalize=True, dropna=True)
joint_prob.to_csv("joint_prob.csv")
```

```
#Marginal Probability Distribution of Petal Length and Sepal Length
#Values between 0 and 1
marginal_prob=pd.crosstab(index=iris_df['petal_len'], columns=iris_df['sepal_len'], margins=True, normalize=True, dropna=True)
marginal_prob.to_csv("marginal_prob.csv")
```

```
#plotting the probability distribution results and saving them as .png images
plt.title("Joint Probability Distribution")
sns.heatmap(joint_prob)
plt.savefig("joint_prob.png")

```

```
plt.title("Marginal Probability Distribution")
sns.heatmap(marginal_prob)
plt.savefig("marginal_prob.png")
```
